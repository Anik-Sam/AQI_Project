```python
# Function to automatically calculate the groupby and plot for that specif column with respect to the city 
def groupby_plot(df5,col):
    df_col = df5.groupby("City")[col].mean().round(2).reset_index()
    df_col = df_col.sort_values(by=col, ascending=False)
    df_col = df_col.reset_index(drop=True)
    
    plt.figure(figsize=(20,15))
    sns.barplot(x="City",y=col,data=df_col)
    plt.xlabel("City")
    plt.ylabel("column")
    plt.title("City Wise data ")
    plt.xticks(rotation=45,ha='right')
    plt.show()
        
    return df_col
```

```python
def groupby_plot(df5,col1,col2):
    df_col = df5.groupby("City")[col1,col2].mean().round(2).reset_index()
    df_col = df_col.sort_values(by=col1, ascending=False)
    df_col = df_col.reset_index(drop=True)
    
    plt.figure(figsize=(20,15))
    sns.scatterplot(x="City",y=col1,hue=col2,data=df_col)
    plt.xlabel("City")
    plt.ylabel(col1)
    plt.title("Scatter Plot Analysis ")
    plt.xticks(rotation=45,ha='right')
    plt.show()
        
    return df_col
```

```python
# Using A function to visualise the Partcile matter and the AQI of any selected city.

def groupby_plot_aqi(df5,City):
    df_location=df5[df5["City"]==City]
    df_location["Year"]=pd.to_datetime(df_location["Date"]).dt.year
    
    df_location_groupby_AQI=df_location.groupby("Year")["AQI"].mean().reset_index()
    df_location_groupby_PM=df_location.groupby("Year")["PM2.5"].mean().reset_index()
    
    plt.figure(figsize=(10,6))
    sns.barplot(x="Year",y="AQI",data=df_location_groupby_AQI)
    plt.xlabel("Year")
    plt.ylabel("AQI")
    plt.title("AQI Analysis ")
    plt.xticks(rotation=45,ha="right")
    
    
    plt.figure(figsize=(10,6))
    sns.barplot(x="Year",y="PM2.5",data=df_location_groupby_PM)
    plt.xlabel("Year")
    plt.ylabel("PM2.5")
    plt.title("Partical Matter Analysis")
    plt.xticks(rotation=45,ha="right")
    
    return df_location_groupby_AQI,df_location_groupby_PM
    
    
    
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
df5["City"].unique()
```

```python
df5 = pd.get_dummies(df5, columns=['City'], drop_first=True)
df5['AQI_Bucket'] = le.fit_transform(df5['AQI_Bucket'])
```

```python
df5.head()
```

```python
df5.columns
```

```python
x=df5[['PM2.5', 'PM10', 'NO', 'NO2', 'NOx', 'NH3', 'CO', 'SO2', 'O3',
       'Benzene', 'Toluene', 'Xylene', 'AQI', 'City_Aizawl',
       'City_Amaravati', 'City_Amritsar', 'City_Bengaluru', 'City_Bhopal',
       'City_Brajrajnagar', 'City_Chandigarh', 'City_Chennai',
       'City_Coimbatore', 'City_Delhi', 'City_Ernakulam', 'City_Gurugram',
       'City_Guwahati', 'City_Hyderabad', 'City_Jaipur', 'City_Jorapokhar',
       'City_Kochi', 'City_Kolkata', 'City_Lucknow', 'City_Mumbai',
       'City_Patna', 'City_Shillong', 'City_Talcher',
       'City_Thiruvananthapuram', 'City_Visakhapatnam']]
y=df5["AQI_Bucket"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import *
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
```

```python
lr=LogisticRegression()
```

```python
model=lr.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
# accuracy sccore of the model 
accuracyscore=accuracy_score(y_test,y_pred)
print(accuracyscore)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier()
```

```python
model_rf=rf.fit(x_train,y_train)
y_predrf=model_rf.predict(x_test)
accuracyscore_rf=accuracy_score(y_test,y_predrf)
print(accuracyscore_rf)
```

```python
# logistic regression and random forest is applied on the whole datatset ,
# the random forest model is giving higher accurate results than logistic regression 
```

```python
y_pred_train=model_rf.predict(x_train)
train_accuracy = accuracy_score(y_train, y_pred_train)
```
